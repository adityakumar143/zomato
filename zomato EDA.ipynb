```python
# 1. Distribution of Aggregate Ratings

plt.figure(figsize=(8, 5))
sns.histplot(df["Aggregate rating"], bins=20, kde=True, color="blue")
plt.title("Distribution of Aggregate Ratings", fontsize=14)
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.show()
```

```python


#Restaurant Ratings: The majority of restaurants have ratings between 2.5 and 4.5, meaning that most restaurants are perceived as average to good. Very few restaurants have extreme ratings (either very low or very high).
```

```python
# 2. Top 10 Cities with Most Restaurants


top_cities = df["City"].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_cities.values, y=top_cities.index, palette="viridis")
plt.title("Top 10 Cities with Most Restaurants", fontsize=14)
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.show()
```

```python
#  City-Wise Restaurant Distribution: Cities like New Delhi, Bangalore, and Mumbai dominate in the number of restaurants, indicating that these are the biggest food hubs.
```

```python
# 3. Price Range Distribution

plt.figure(figsize=(7, 5))
sns.countplot(x=df["Price range"], palette="coolwarm")

plt.title("Distribution of Price Ranges", fontsize=14)
plt.xlabel("Price Range")
plt.ylabel("Count")
plt.show()
```

```python
#   Affordability: Most restaurants fall under price ranges 1 and 2, meaning that budget-friendly dining options are more prevalent compared to high-end restaurants
```

```python
# 4. Top 10 Most Common Cuisines

top_cuisines = df["Cuisines"].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette="magma")
plt.title("Top 10 Most Common Cuisines", fontsize=14)
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine Type")
plt.show()
```

```python
#   Popular Cuisines: North Indian, Chinese, and Fast Food are the most commonly available cuisines, showing that these food preferences are widely accepted.
```

```python
# 5. Ratings vs. Number of Votes

plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Aggregate rating"], y=df["Votes"], alpha=0.5)
plt.title("Aggregate Rating vs. Number of Votes", fontsize=14)
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Votes")
plt.show()
```

```python
#   Customer Engagement: Restaurants with higher ratings tend to receive more votes, indicating that well-rated places attract more customer reviews and engagement.
```

```python
# 6. Online Delivery Availability

plt.figure(figsize=(6, 4))
sns.countplot(x=df["Has Online delivery"], palette="pastel")
plt.title("Online Delivery Availability", fontsize=14)
plt.xlabel("Has Online Delivery")
plt.ylabel("Count")
plt.show()
```

```python
#  Online Delivery Trends: A significant number of restaurants do not offer online delivery, which suggests either a preference for dine-in experiences or operational constraints.
```

```python
# 7. Table Booking Availability

plt.figure(figsize=(6, 4))
sns.countplot(x=df["Has Table booking"], palette="muted")
plt.title("Table Booking Availability", fontsize=14)
plt.xlabel("Has Table Booking")
plt.ylabel("Count")
plt.show()
```

```python
# Table Booking System: Many restaurants operate without table booking services, which may indicate that casual dining and fast food places dominate over fine dining.
```

```python
# 8. Distribution of Votes

plt.figure(figsize=(8, 5))
sns.histplot(df["Votes"], bins=30, kde=True, color="green")
plt.title("Distribution of Votes", fontsize=14)
plt.xlabel("Number of Votes")
plt.ylabel("Frequency")
plt.show()
```

```python
#  Popularity and Votes: Most restaurants receive a low number of votes, meaning that only a few highly popular places dominate customer attention and engagement.
```

```python
# 9. Ratings by Price Range

plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Price range"], y=df["Aggregate rating"], palette="coolwarm")
plt.title("Aggregate Ratings by Price Range", fontsize=14)
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
#  Price vs. Ratings: Restaurants with higher price ranges generally receive better ratings, suggesting that customers may associate higher prices with better food quality and service.
```

```python
# 10. Average Cost for Two Distribution

plt.figure(figsize=(8, 5))
sns.histplot(df["Average Cost for two"], bins=30, kde=True, color="purple")
plt.title("Distribution of Average Cost for Two", fontsize=14)
plt.xlabel("Average Cost for Two")
plt.ylabel("Frequency")
plt.xlim(0, df["Average Cost for two"].quantile(0.95))  # Limit extreme values
plt.show()
```

```python
# Average Cost for Two: The majority of restaurants have a reasonable cost for two people, though a few high-end establishments cater to premium customers.
```

```python
# Top 10 Restaurants Based on Aggregate Rating

top_restaurants = df[['Restaurant Name', 'Aggregate rating']].sort_values(
    by='Aggregate rating', ascending=False).drop_duplicates().head(10)

# Visualization

plt.figure(figsize=(10, 5))
sns.barplot(y=top_restaurants['Restaurant Name'], x=top_restaurants['Aggregate rating'], palette="Blues_r")
plt.title("Top 10 Restaurants Based on Aggregate Rating", fontsize=14)
plt.xlabel("Aggregate Rating")
plt.ylabel("Restaurant Name")
plt.xlim(4, 5)  # Focus on highly-rated restaurants
plt.show()
```
